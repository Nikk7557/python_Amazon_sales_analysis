```python
dy=sns.countplot(x="Size",data=df)
```

Note; from above graph you can see that most of the people but M size

**Group By**

The group buy function in pandas is used to  group data based on one or more columns in a DataFrame

```python
df.groupby(["Size"],as_index=False)["Qty"].sum().sort_values(by="Qty",ascending=False)
```

```python
P_Qty=df.groupby(["Size"],as_index=False)["Qty"].sum().sort_values(by="Qty",ascending=False)
sns.barplot(x="Size",y="Qty",data=P_Qty)
```

Note; from the above graph you can see that most of the qty buys M size in the sales

**Courier Status**

```python
sns.countplot(data=df,x="Courier Status",hue="Status")
```

```python
plt.figure(figsize=(10,5))
dy=sns.countplot(data=df,x="Courier Status",hue="Status")
plt.show()
```

Note; from the above graph you can see that the majority of the orders are shipped by the couriers

```python
#histogram
df["Category"]=df["Category"].astype(str)
column_data=df["Category"]
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=20,edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

from the above graph you can see most of the buyers are T-shirt

```python
#checking b2b Data by using pie chart
B2B_check=df['B2B'].value_counts()
#plot the pie chart
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
#plt.axis(equal)
plt.show()
```

from the above chart we can see that maximum i,e. 99% of buyers are retailers and 1%  are B2B buyers

```python
#prepare data for scater plot
x_data=df['Category']
y_data=df['Size']
#plot the scater plot
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('scatter Plot')
plt.show()
```

```python
#top 10 states
top_10_states=df['ship-city'].value_counts().head(10)
#plot count of cities  by state
plt.figure(figsize=(10,6))
sns.countplot(data=df[df['ship-city'].isin(top_10_states.index)],x='ship-state')
plt.xlabel('ship-city')
plt.ylabel('count')
plt.title('distribution of state')
plt.xticks(rotation=45)
plt.show()
```

from the above graph most of the buyers are fron maharashtra state

**Conclusion** : The Data Analysis  reveals that business has a significant customer based in Maharastra state many serves retailers,fulfills orders through Amazon,experience high demand for T-shirts and see M size as the prefered choice among buyers
